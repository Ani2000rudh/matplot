```python
#Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated dataplot a scatter plot.

"scatterplot just marks the (x,y) points and does not connect it with line"
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
pt.scatter(x,y)
pt.xlabel('this is x axis')
pt.ylabel('this is y axis')
pt.title('this is 2nd question')
pt.show()
```

```python
'''Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])'''


'''A bar graph is a type of chart used to visually represent 
data using rectangular bars. The bars can be either horizontal or vertical, 
and their lengths are proportional to the data they represent.'''
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
pt.bar(company,profit)
```

```python
'''Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)'''


'''A box plot, also known as a box and whisker
plot, is a type of chart used to represent a summary of a dataset. 
It provides a visual representation of the minimum, maximum, median, first quartile, and third quartile of the data,
as well as any outliers or extreme values.'''

# create the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# combine the data into a list
data = [box1, box2]

# create the box plot
fig, ax = plt.subplots()
ax.boxplot(data)

# set the labels for the x-axis and y-axis
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')

# display the plot
plt.show()
```
